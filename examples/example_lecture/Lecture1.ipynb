This is an example lecture. We have some data in a file called `somedata.csv` that we wish to plot.  Here's how we might do it in Python

```python
# We'll use pandas to load the data as a dataframe
import pandas as pd
df = pd.read_csv('somedata.csv', sep=',',header=None)
```

```python
# Plot using matplotlib
import matplotlib
%matplotlib inline
df.plot(x=0,y=1,style='r--o');
```
